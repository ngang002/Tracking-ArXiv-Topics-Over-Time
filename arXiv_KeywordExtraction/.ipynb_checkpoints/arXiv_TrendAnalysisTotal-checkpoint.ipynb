# Modules for Trend Analysis

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w


def autoCorrelate(series):
    return signal.correlate(series, series, mode='same') # / len(series)


def fourier_analysis(series, sampling_interval=24):
    series /= np.sum(series)
    N = len(series)
    yf = fft(series) #-series.mean())
    xf = fftfreq(N, d=sampling_interval)
    return xf[:N//2], np.abs(yf[:N//2])  # Only positive frequencies
```

```python
# Structuring the time-series data as [TOPIC --> [TIME SERIES DATA]] --> SHAPE: [NTOPICS, NMONTHS]
YEAR_START, YEAR_END = 1995, 2022
timeLine = np.zeros(len(range(YEAR_START, YEAR_END))*len(range(1, 13)))

dates = []
# dates = df_FirstTopic.groupby(['Year', 'Month'])['Year'].unique()
# print(dates)
# Counting the number of times a topic appears in a month, and saving it to the timeLine array
for YY, YEAR in enumerate(range(YEAR_START, YEAR_END)):
    for MM, MONTH in enumerate(range(1, 13)):
        # if YY*12+MM >= 175: break

        dates.append([[MONTH,YEAR]])
        KEEP_IND = np.logical_and(df_FirstTopic['Year'] == YEAR, df_FirstTopic['Month'] == MONTH)
        timeLine[YY*12+MM] = len(df_FirstTopic[KEEP_IND])
```

## Covariance Between Months

```python
def covariance2D(timeSeries):
    meanTS = np.mean(timeSeries)
    covariance = np.zeros((len(timeSeries),len(timeSeries)))
    for i in range(len(timeSeries)):
        for j in range(i,len(timeSeries)):
            _ = (timeSeries[i]-meanTS)*(timeSeries[j]-meanTS)
            covariance[i,j] = _
            covariance[j,i] = _

    return covariance
    

fig, ax = plt.subplots(figsize=(10,5), ncols=2)
cov2D = covariance2D(additiveDecompositions['observed'])# np.cov(additiveDecompositions['trend'][120:125])
ax[0].set_title("Covariance Matrix of Observed")
im = ax[0].imshow(cov2D, cmap='rainbow', vmin=-100000, vmax=100000)

cov2D = covariance2D(additiveDecompositions['trend']) # np.cov(additiveDecompositions['trend'][120:125])
ax[1].set_title("Covariance Matrix of Trend")
im = ax[1].imshow(cov2D, cmap='rainbow', vmin=-100000, vmax=100000)
fig.colorbar(im, ax=ax, orientation='vertical', fraction=.1)
```

In the above data we can see that largely the covariance matrix is close to 0 near the center of the time-series data, while the covariation increases at the edges of the data.  In my initial visual analysis of the trend in the data I noticed that early on in the data gathering the number of papers increases over time, up to about the 100th time-step.  Meanwhile, at around the 150th timestep the data begins to flatten a bit more, where the slope of the increase is substantially less than before.  Additioanlly, we being to notice noticeable (but seemingly periodic) dips during this period as well.  However, this flattening can also be causing the large covariances at later times as well, while also showing random anti-covariances in the real observed data. 

# Fitting the Trend 

## Laying Out Models

In this section, I am going to test multiple models of growth to predict the trend.  Remember, there are two separate aspects of our predictive models: the trend and the periodicity in papers.  In this section I will first model the trends in the number of papers published over time on the arXiv. In order to accomplish this goal I will test 3 different models (although others can be used, I believe I have chosen the simplest and most intuitive options).  

1. Linear Growth  --> $N(t) = mt+b$


2.  Logistic Growth  --> $N(t) = \frac{a}{1+b^{t}}+c$


3. Power Law Growth --> $N(t) = ax^{t}+c$


These options were largely picked, because while simple, they are also extremely powerful predictors, and sometimes the simplest model can best explain the data.  

```python
from scipy.optimize import curve_fit
import statsmodels.api as sm

def linearGrowth(x, a, b):
    return a*x+b

def logisticGrowth(x, a, b, c):
    return a/(1.+b**(-x))+c

def powerGrowth(x, a, b, c):
    return a*np.power(x, b)+c
```

In the cells below I will now fit the data to each of these fits (although I will note that the linear model will bre fit using the statsmodels Generalized Least Squares method.  Additionally, and VERY importantly since our analysis is being done with time-series data, it will be necessary to calculate the covariance matrix since the observations during some month *m* can impact the number of papers observed in month *m+1*.

## Fitting the Lines

```python
stepsFit = steps
stepsFitExtended = np.concatenate([-steps, steps])
lineFit = additiveDecompositions['trend']
lineFitExtended = np.concatenate([-additiveDecompositions['trend'], additiveDecompositions['trend']])

# Linear model
t = sm.add_constant(steps)  # adds intercept term
model = sm.GLS(lineFit, t, disp=0).fit()
m, b = model.params[1], model.params[0]
print(f"Linear Growth Function: y(t)={m}*t+{b}")
linGrowth_cov = model.cov_params()
print(f"Conditional Number: {np.linalg.cond(linGrowth_cov)}")

# Logistical Growth Model
# 3 Variables
logGrowth_t, logGrowth_cov = curve_fit(logisticGrowth, xdata=stepsFitExtended, ydata=lineFitExtended, p0=[0., 1., stepsFit[0]])
print(f"Logistic Growth Functions: y(t)={logGrowth_t[0]}/(1+{logGrowth_t[1]}^x)+{logGrowth_t[2]}")
# 2 Variables
# logGrowth_t, logGrowth_cov = curve_fit(logisticGrowth, xdata=stepsFitExtended, ydata=lineFitExtended, p0=[0., 1.])
# print(f"Logistic Growth Functions: y(t)={logGrowth_t[0]}/(1+{logGrowth_t[1]}^x)")
print(f"Conditional Number: {np.linalg.cond(logGrowth_cov)}")

# Power Law Growth
# 3 Variables 
powerGrowth_t, powerGrowth_cov = curve_fit(powerGrowth, xdata=steps, ydata=lineFit, p0=[1., 1.0, stepsFit[0]], method='trf')
print(f"Power Law Function: y(t)={powerGrowth_t[0]}*x^{powerGrowth_t[1]})+{powerGrowth_t[2]}")
# 2 Varialbes
# powerGrowth_t, powerGrowth_cov = curve_fit(powerGrowth, xdata=steps, ydata=lineFit, p0=[1., 0.5], method='trf')
# print(f"Power Law Function: y(t)={powerGrowth_t[0]}*x^{powerGrowth_t[1]})")
# 1 Varialbes
# powerGrowth_t, powerGrowth_cov = curve_fit(powerGrowth, xdata=steps, ydata=lineFit, method='trf')
# print(f"Power Law Function: y(t)=x^{0.5}+{powerGrowth_t[0]}")
print(f"Conditional Number: {np.linalg.cond(powerGrowth_cov)}")
```

```python
fig, ax = plt.subplots(figsize=(15,5))

yLinear   = linearGrowth(steps, *[m,b])
yLogistic = logisticGrowth(steps, *logGrowth_t)
yPower    = powerGrowth(steps, *powerGrowth_t)

ax.scatter(steps, additiveDecompositions['trend'], c='k', label='Original Trend Data')
ax.plot(steps, yLinear,    c='r', lw=3., label='Linear function')
ax.plot(steps, yLogistic,  c='b', lw=3., label='Logistic function')
ax.plot(steps, yPower ,    c='g', lw=3., label='Root Power function')
ax.legend(loc='best')
ax.set_xlabel("Time (Months)")
ax.set_ylabel("Number of Papers From Trend")

plt.show()

fig, ax = plt.subplots(figsize=(12,2), ncols=3)
# Loop over data dimensions and create text annotations.

ax[0].imshow(linGrowth_cov,)
for i in range(len(linGrowth_cov)): 
    for j in range(len(linGrowth_cov[0])):  
        text = ax[0].text(j, i, f"{linGrowth_cov[i, j]:0.4f}", ha="center", va="center", color="w")
ax[1].imshow(logGrowth_cov,)
for i in range(len(logGrowth_cov)): 
    for j in range(len(logGrowth_cov[0])):  
        text = ax[1].text(j, i, f"{logGrowth_cov[i, j]:0.4f}", ha="center", va="center", color="w")
ax[2].imshow(powerGrowth_cov,)
for i in range(len(powerGrowth_cov)): 
    for j in range(len(powerGrowth_cov[0])):  
        text = ax[2].text(j, i, f"{powerGrowth_cov[i, j]:0.4f}", ha="center", va="center", color="w")
fig.suptitle("Covariance Matrices",fontsize=12)
plt.show()
```

In the first plot above, I show the results of the best fit linear, logistic, and power law models, and through simple visual inspection it appears that the power law is the best fit line.  However, our final conclusion can only be made after some further investigation.  However, if we examine the covariance matrix we can start building a clearer picture.  Our linear model has only two free parameters in our model, whereas the other 2 have 3.  The slope of the line seemingly has a relatively large variance, while the intercept has a much smaller variance.  However, for the toher two models, we notice something interesting.  While the exponential term (b) in both of our models have very small variances, the variances for a and c are substantially much larger (and the covariance between them is also quite large).  This is because (in both models) we can vary these variables, and likely only one of these needs to be selected.  However, after further examination, using only 2 variables the lines that are best fit are very poor fits (leading to absurdly large residuals).  In short, any changes made for a that causes deviation from the best fit line, can be modulated by c (and vice-versa).  However, without both scipy.optimize.curve_fit will not adequately fit for the given data.

# Fitting the Periodic Data

```python
import pmdarima
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
print(pmdarima.__version__)
```

## Testing Stationary Assumption for Periodic Data

```python
# Let us simply set the period data to a new variable ts 
ts = additiveDecompositions['seasonal']

dftest = adfuller(ts, autolag="AIC")
dfoutput = pd.Series(
    dftest[0:4],
    index=[
        "Test Statistic",
        "p-value",
        "#Lags Used",
        "Number of Observations Used",
    ]
)
print(dfoutput)
```

```python
# Fit an auto ARIMA model
bestModel = auto_arima(
    ts,
    start_p=1, max_p=6,
    start_q=1, max_q=6,
    d=None, # will be done based on KPSS tests
    seasonal=True,
    start_P=0, max_P=6,
    start_Q=0, max_Q=6,
    D=1,
    m=12,  # Seasonal period (e.g., 12 for monthly data)
    trace=True,  # Print search progress
    error_action='ignore',
    suppress_warnings=True,
    stepwise=False,  # Stepwise search to reduce computation time 
    n_jobs = -1 # 
)

BEST_MODEL = {'Order': bestModel.order, 'SeasonalOrder': bestModel.seasonal_order}
```

```python
bestModelOrders= BEST_MODEL
order          = bestModelOrders['Order']
seasonalOrder  = bestModelOrders['SeasonalOrder']

# order, seasonalOrder = (1, 1, 1), (1, 1, 2, 12)
seasonalModel = SARIMAX(ts, order=order, seasonal_order=seasonalOrder, simple_differencing=False, enforce_stationarity=False)
result = seasonalModel.fit(disp=False)
predEvolution = result.forecast(len(ts))

fig, ax = plt.subplots(figsize=(12, 4),nrows=2)
ax[0].plot(np.arange(len(predEvolution)), predEvolution, '-b')
ax[0].plot(np.arange(len(ts)), additiveDecompositions['seasonal'], '--r')

ax[0].set_ylabel("Number of Papers from Seasonality")
ax[0].set_xlabel('Time (Months)')

ax[1].plot(np.arange(len(predEvolution)), predEvolution-ts, '-k')
ax[1].set_xlabel('Time (Months)')
ax[1].set_ylabel('Residuals')


fig.show()

# result.plot_diagnostics(figsize=(10, 8))
```
